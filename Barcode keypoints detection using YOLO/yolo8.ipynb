## Measuring normalized Hausdorff distance

```python
import os
import numpy as np
from scipy.spatial.distance import directed_hausdorff

# Directories for predicted and ground truth data
pred_directory = "/home/aous/Desktop/MIPT/project/yolo data/data splitted/predicted/26"
gt_directory = "/home/aous/Desktop/MIPT/project/yolo data/data splitted/test"

# Counters and lists for results
count_of_correct_kp_number = 0
count_of_incorrect_kp_number = 0
normalized_hausdorff_list = []
hausdorff_list = []
list_of_names = []
list_of_missing_or_additional_kp = []
list_of_false_number = []
t = []
t2 = []
diagonla_bb = []

# Function to extract keypoints and normalize based on bounding box size
def extract_and_normalize_keypoints(values):
    values = [float(val) for val in values]
    values = values[5:]  # Skipping the first five values (class label and bounding box details)
    if values[-1] == 0:  # Removing any trailing zero that might indicate an empty keypoint
        values = values[:-6]
    values = [val * 640 for val in values if val != 2]
    keypoints = np.array([values[i:i+2] for i in range(0, len(values), 2)])
    
    # keypoints *= 640  # Scaling the keypoints (assuming normalization to a 640x640 image)
    return keypoints

# Iterating through ground truth files
for filename in os.listdir(gt_directory):
    if filename.endswith(".txt"):
        list_of_names.append(filename)
        
        # Reading ground truth keypoints
        with open(os.path.join(gt_directory, filename), 'r') as file:
            values = file.read().split()
        bbox_width = float(values[3]) * 640  # Width of the bounding box
        bbox_height = float(values[4]) * 640  # Height of the bounding box
        gt_keypoints = extract_and_normalize_keypoints(values)
        t.append(len(gt_keypoints))
        
        # Reading predicted keypoints
        with open(os.path.join(pred_directory, filename), 'r') as file:
            values = file.read().split()
        pred_keypoints = np.array([int(val) for val in values]).reshape(-1, 2)
        t2.append(len(pred_keypoints))
        
        # Check if the number of keypoints matches
        if len(gt_keypoints) == len(pred_keypoints):
            # Calculate Hausdorff distance
            distance_0 = directed_hausdorff(gt_keypoints, pred_keypoints)[0]
            distance_1 = directed_hausdorff(pred_keypoints, gt_keypoints)[0]
            hausdorff_distance = max(distance_0, distance_1)
            
            # Normalize the Hausdorff distance
            normalized_hausdorff_distance = hausdorff_distance / np.sqrt(bbox_width**2 + bbox_height**2)
            hausdorff_list.append(hausdorff_distance)
            diagonla_bb.append(np.sqrt(bbox_width**2 + bbox_height**2))
            list_of_missing_or_additional_kp.append(0)
            normalized_hausdorff_list.append(normalized_hausdorff_distance)
            count_of_correct_kp_number += 1
        else:
            d = len(pred_keypoints) - len(gt_keypoints)
            list_of_false_number.append(filename)
            list_of_false_number.append(len(pred_keypoints))
            list_of_missing_or_additional_kp.append(d)
            list_of_false_number.append(pred_keypoints)
            count_of_incorrect_kp_number += 1


normalized_hausdorff_list = [round(num, 4) for num in normalized_hausdorff_list]
hausdorff_list = [round(num, 2) for num in hausdorff_list]
diagonla_bb = [round(num) for num in diagonla_bb]
# Output results
print("Count of correct keypoints number:", count_of_correct_kp_number)
print("Count of incorrect keypoints number:", count_of_incorrect_kp_number)
print("Hausdorff distances (normalized):", normalized_hausdorff_list[:4])
print("Hausdorff distances:", hausdorff_list[:4])
print("Diagonal BB list:", diagonla_bb[:4])
```

```python
import os
import numpy as np
from scipy.spatial.distance import directed_hausdorff

# Directories for predicted and ground truth data
pred_directory = "/home/aous/Desktop/MIPT/project/yolo data/data splitted/predicted/20"
gt_directory = "/home/aous/Desktop/MIPT/project/yolo data/data splitted/test"

# Counters and lists for results
count_of_correct_kp_number = 0
count_of_incorrect_kp_number = 0
normalized_hausdorff_list = []
hausdorff_list = []
list_of_names = []
list_of_missing_or_additional_kp = []
list_of_false_number = []
t = []
t2 = []
diagonla_bb = []

# Function to extract keypoints and normalize based on bounding box size
def extract_and_normalize_keypoints(values):
    values = [float(val) for val in values]
    values = values[5:]  # Skipping the first five values (class label and bounding box details)
    if values[-1] == 0:  # Removing any trailing zero that might indicate an empty keypoint
        values = values[:-6]
    values = [val * 640 for val in values if val != 2]
    keypoints = np.array([values[i:i+2] for i in range(0, len(values), 2)])
    
    # keypoints *= 640  # Scaling the keypoints (assuming normalization to a 640x640 image)
    return keypoints

# Iterating through ground truth files
for filename in os.listdir(gt_directory):
    if filename.endswith(".txt"):
        list_of_names.append(filename)
        
        # Reading ground truth keypoints
        with open(os.path.join(gt_directory, filename), 'r') as file:
            values = file.read().split()
        bbox_width = float(values[3]) * 640  # Width of the bounding box
        bbox_height = float(values[4]) * 640  # Height of the bounding box
        gt_keypoints = extract_and_normalize_keypoints(values)
        t.append(len(gt_keypoints))
        
        # Reading predicted keypoints
        with open(os.path.join(pred_directory, filename), 'r') as file:
            values = file.read().split()
        pred_keypoints = np.array([int(val) for val in values]).reshape(-1, 2)
        t2.append(len(pred_keypoints))
        
        # Check if the number of keypoints matches
        if len(gt_keypoints) == len(pred_keypoints):
            # Calculate Hausdorff distance
            distance_0 = directed_hausdorff(gt_keypoints, pred_keypoints)[0]
            distance_1 = directed_hausdorff(pred_keypoints, gt_keypoints)[0]
            hausdorff_distance = max(distance_0, distance_1)
            
            # Normalize the Hausdorff distance
            normalized_hausdorff_distance = hausdorff_distance / np.sqrt(bbox_width**2 + bbox_height**2)
            hausdorff_list.append(hausdorff_distance)
            diagonla_bb.append(np.sqrt(bbox_width**2 + bbox_height**2))
            list_of_missing_or_additional_kp.append(0)
            normalized_hausdorff_list.append(normalized_hausdorff_distance)
            count_of_correct_kp_number += 1
        else:
            d = len(pred_keypoints) - len(gt_keypoints)
            list_of_false_number.append(filename)
            list_of_false_number.append(len(pred_keypoints))
            list_of_missing_or_additional_kp.append(d)
            list_of_false_number.append(pred_keypoints)
            count_of_incorrect_kp_number += 1


normalized_hausdorff_list = [round(num, 4) for num in normalized_hausdorff_list]
hausdorff_list = [round(num, 2) for num in hausdorff_list]
diagonla_bb = [round(num) for num in diagonla_bb]
# Output results
print("Count of correct keypoints number:", count_of_correct_kp_number)
print("Count of incorrect keypoints number:", count_of_incorrect_kp_number)
print("Hausdorff distances (normalized):", normalized_hausdorff_list[:4])
print("Hausdorff distances:", hausdorff_list[:4])
print("Diagonal BB list:", diagonla_bb[:4])
```

```python
iou_values = []
for i in iou_list:
    box_gt = i[1]
    box_pred = i[0]
    x1_min, y1_min = box_pred[0] - box_pred[2]/2, box_pred[1] - box_pred[3]/2
    x1_max, y1_max = box_pred[0] + box_pred[2]/2, box_pred[1] + box_pred[3]/2

    x2_min, y2_min = box_gt[0] - box_gt[2]/2, box_gt[1] - box_gt[3]/2
    x2_max, y2_max = box_gt[0] + box_gt[2]/2, box_gt[1] + box_gt[3]/2

    intersection_area = max(0, min(x1_max, x2_max) - max(x1_min, x2_min)) * max(0, min(y1_max, y2_max) - max(y1_min, y2_min))
    union_area = box_pred[2] * box_pred[3] + box_gt[2] * box_gt[3] - intersection_area
    IoU = intersection_area / union_area
    iou_values.append(IoU)
IoU_average = sum(iou_values) / 628
print(f"Average IoU for Bounding Box is: {IoU_average}")
```
